# Answering questions about a police report
In this lesson, we will send a police report to ChatGPT and then ask questions about it

## Setup

````python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

openai.api_key  = os.getenv('OPENAI_API_KEY')
````

````python
def get_completion(prompt, model="gpt-4"): 
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

## Summarize the first fragment

````python
prompt = f"""
Your task is to generate a short summary of the report.

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Report: ```{report_fragment_1}```
"""

response = get_completion(prompt)
print(response)
````

## Ask questions

````python
prompt = f"""
Your task is to answer my question based on the report fragment

Report: ```{report_fragment_1}```
Question: Is this charge a wobbler under the law in California
"""

response = get_completion(prompt)
print(response)
````

## Summarize multiple fragments

````python
report_fragments = [report_fragment_1, report_fragment_2]
````

````python
for i in range(len(report_fragments)):
    prompt = f"""
    Your task is to generate a short summary of a product \ 
    review from an ecommerce site. 

    Summarize the review below, delimited by triple \
    backticks in at most 20 words. 

    Review: ```{report_fragments[i]}```
    """

    response = get_completion(prompt)
    print(i, response, "\n")
````
