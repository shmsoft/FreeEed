```python
def break_into_pages(filename):
    with open(filename, 'r') as file:
        content = file.read()

    # pages = content.split('\f')

    pages = content.split("\n\n")
    
    return pages
```

```python
def write_pages_to_files(pages, filename_prefix, output_dir):
    for i, page in enumerate(pages):
        with open(f"{output_dir+filename_prefix}_page_{i+1}.txt", 'w') as file:
            file.write(page)
```

```python
import re

def break_into_paragraphs(filename):
    with open(filename, 'r') as file:
        document = file.read()
    lines = document.split('\n')

    paragraphs = []
    current_paragraph = []

    for line in lines:
        if re.match(r'^\d+$', line.strip()):
            if current_paragraph:
                paragraphs.append('\n'.join(current_paragraph))
                current_paragraph = []
        else:
            current_paragraph.append(line)

    if current_paragraph:
        paragraphs.append('\n'.join(current_paragraph))
    
    return paragraphs
```
