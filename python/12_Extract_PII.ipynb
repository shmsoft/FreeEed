```python
import boto3
```

```python
comprehend = boto3.client(service_name='comprehend', region_name='us-east-2')
```

```python
text = "Your text with potential PII here. My name is John Doe. I live at 123 Main"
```

```python
response = comprehend.detect_pii_entities(Text=text, LanguageCode='en')
```

```python
pii_entities = response['Entities']
for entity in pii_entities:
    print(f"Entity: {entity['Type']}, Score: {entity['Score']}, BeginOffset: {entity['BeginOffset']}, EndOffset: {entity['EndOffset']}")
```

```python
pii_entities = response['Entities']
for entity in pii_entities:
    entity_text = text[entity['BeginOffset']:entity['EndOffset']]
    print(f"Entity: {entity['Type']}, Score: {entity['Score']}, Text: \"{entity_text}\"")
```
