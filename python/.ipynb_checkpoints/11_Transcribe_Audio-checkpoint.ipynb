# Transcribe audio file with AssemblyAI

```python
import os
from dotenv import load_dotenv, find_dotenv
import assemblyai as aai
_ = load_dotenv(find_dotenv()) # read local .env file
aai.settings.api_key = os.getenv('ASSEMBLY_AI_KEY')
```

```python
transcriber = aai.Transcriber()
```

```python
audio_file = "../../data/7510.mp3"
transcript = transcriber.transcribe(audio_file)
```

```python
print(transcript.text)
```

```python
print(transcript.words)
```

```python
sentences = transcript.get_sentences()
for sentence in sentences:
  print("sentence.text=" + sentence.text)
  print("sentence.words=" + str(sentence.words))

paragraphs = transcript.get_paragraphs()
for paragraph in paragraphs:
  print("paragraph.text=" + paragraph.text)
```
