# Playground to store a document coming in as text in a Pinecone index with namespaces

```python
# document_path # coming as parameter
document_path = "../../data/Court Documents Probation Pre Sentence Report Vasquez.txt"
!ls ../../data
```

```python
with open(document_path, 'r') as file:
    document_content = file.read()
```

```python
print(len(document_content))
```

```python
import os
import openai
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain_pinecone import Pinecone
import pinecone
```

```python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file
openai.api_key  = os.getenv('OPENAI_API_KEY')
```

```python
index_name = os.getenv('PINECONE_INDEX_NAME')
model = os.getenv('LLM_MODEL')
```

```python
namespace = "01_Store_document"
```

```python
temp_file = "/tmp/temp.txt"
with open(temp_file, 'w') as f:
    # Write the string to the file
    f.write(document_content)
```

```python
pc = Pinecone(api_key=os.getenv('PINECONE_API_KEY'))
```

```python
# def split_docs(documents, chunk_size=1000, chunk_overlap=20):
def split_docs(documents, chunk_size=10, chunk_overlap=10):
    # text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    text_splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    docs = text_splitter.split_documents(documents)
    return docs
```

```python
loader = TextLoader(document_path)
documents = loader.load()
```

```python
docs = split_docs(documents)
```

```python
print(len(docs)) # log trace
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=openai.api_key)
```

```python
Pinecone.from_documents(docs, embeddings, index_name=index_name, namespace = namespace)
```
