```python
import os
import openai
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Pinecone
import pinecone
```

```python
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file
openai.api_key  = os.getenv('OPENAI_API_KEY')
```

```python
index_name = "dl-ai-uadivioxt3blbkfjdpab5b0fd7zfvlidxbsn"
MODEL = "gpt-4"
```

```python
namespace = "159" # coming as parameter
```

```python
temp_file = "/tmp/temp.txt"
with open(temp_file, 'w') as f:
    # Write the string to the file
    f.write(document_content)
```

```python
pinecone.init(
    api_key=os.getenv('PINECONE_API_KEY'),     
    environment="us-west-2"  # next to api key in console
)
```

```python
# def split_docs(documents, chunk_size=1000, chunk_overlap=20):
def split_docs(documents, chunk_size=10, chunk_overlap=10):
    # text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    text_splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    docs = text_splitter.split_documents(documents)
    return docs
```

```python
loader = TextLoader(document_path)
documents = loader.load()
```

```python
docs = split_docs(documents)
```

```python
print(len(docs)) # log trace
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=openai.api_key)
```

```python
Pinecone.from_documents(docs, embeddings, index_name=index_name, namespace = namespace)
```
